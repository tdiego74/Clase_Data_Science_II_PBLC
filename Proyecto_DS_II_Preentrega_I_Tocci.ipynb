#**Proyecto Calidad Vino Tinto**


## **Abstract**
El análisis de la calidad del vino tinto es esencial para entender cómo las propiedades fisicoquímicas del vino influyen en su percepción de calidad. El dataset "Wine Quality" contiene información sobre varios atributos químicos de muestras de vino tinto, así como una calificación de calidad asignada por expertos en una escala de 0 a 10. Este proyecto tiene como objetivo explorar las relaciones entre estas características y la calidad del vino, identificar patrones significativos y formular hipótesis que puedan ser útiles para productores y consumidores de vino.

## **Objetivo del Proyecto**
Este proyecto tiene como objetivo identificar los factores fisicoquímicos que influyen en la calidad del vino tinto, explorar las relaciones entre estas características y la calidad del vino, identificar patrones significativos y proporcionar recomendaciones prácticas que puedan ser útiles para productores y consumidores de vino con el fin de mejorar la calidad del producto.



## **Contexto Comercial**

El mercado del vino es altamente competitivo y los consumidores están cada vez más interesados en vinos de alta calidad. Comprender los factores que afectan la calidad del vino puede ayudar a los productores a ajustar sus procesos de producción para satisfacer mejor las expectativas del mercado y aumentar sus ventas.


## **Contexto Analítico**

Utilizando técnicas de análisis de datos y visualización, este estudio explora las relaciones entre varias características fisicoquímicas del vino tinto y su calidad. Los análisis realizados incluyen gráficos univariados, bivariados y multivariados para identificar patrones y tendencias significativas.


##**Dataset:** : winequality-red.csv

##**Descripción de la temática de los datos:**

El dataset proporcionado es sobre la calidad del vino tinto, que típicamente incluye variables relacionadas con sus características fisicoquímicas y una variable objetivo que indica la calidad del vino.

##**Descripción de las variables:**

**Acidez fija:** Componente principal de los ácidos en el vino que contribuye a su sabor y estabilidad.

**Acidez volátil:** Ácidos que contribuyen a un sabor avinagrado, influenciando negativamente la calidad si están en niveles altos.

**Ácido cítrico:** Contribuye a la frescura y a un sabor ácido en el vino.

**Azúcar residual:** Cantidad de azúcar que queda después de la fermentación, influenciando el dulzor del vino.

**Cloruros:** Contribuyen a la sensación de salinidad en el vino.

**Dióxido de azufre libre:** Previene el crecimiento microbiano y la oxidación.

**Dióxido de azufre total:** Suma del dióxido de azufre libre y el combinado, es crucial para la preservación del vino.

**Densidad:** Influencia el cuerpo del vino y está relacionado con el contenido de alcohol y azúcar.

**pH:** Medida de acidez/alcalinidad del vino, influye en la estabilidad y el sabor.

**Sulfatos:** Contribuyen a la conservación y ayudan a incrementar la intensidad del sabor.

**Alcohol:** Influencia el cuerpo, sabor y estabilidad del vino.

**Calidad:** Variable objetivo que mide la calidad del vino en una escala (generalmente de 0 a 10).

### **Visualización de correlaciones**

```python
# Importar bibliotecas necesarias
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionar solo las columnas numéricas del DataFrame
df_numerico = df.select_dtypes(include='number')

# Calcular la matriz de correlación solo con las columnas numéricas
correlation_matrix = df_numerico.corr()

# Crear el heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linecolor='white', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.grid(False)
plt.show()
```

```python
# Importar bibliotecas necesarias
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar datos de Vino Tinto y Vino Blanco
vino_tinto = df[df['Categoria'] == 'Vino Tinto']
vino_blanco = df[df['Categoria'] == 'Vino Blanco']

# Seleccionar solo las columnas numéricas para calcular las correlaciones
vino_tinto_numerico = vino_tinto.select_dtypes(include='number')
vino_blanco_numerico = vino_blanco.select_dtypes(include='number')

# Calcular las matrices de correlación para cada tipo de vino
correlation_tinto = vino_tinto_numerico.corr()
correlation_blanco = vino_blanco_numerico.corr()

# Crear una figura con dos subplots (uno para cada tipo de vino)
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# Heatmap para Vino Tinto
sns.heatmap(correlation_tinto, annot=True, cmap='coolwarm', linecolor='white', linewidths=0.5, ax=axes[0])
axes[0].set_title('Matriz de Correlación - Vino Tinto')

# Heatmap para Vino Blanco
sns.heatmap(correlation_blanco, annot=True, cmap='coolwarm', linecolor='white', linewidths=0.5, ax=axes[1])
axes[1].set_title('Matriz de Correlación - Vino Blanco')

# Ajustar el diseño y mostrar los gráficos
plt.tight_layout()
plt.show()

```

### **Observaciones Basadas en la Matriz de Correlación**


1. **Correlación entre Alcohol y Calidad:**
   - Existe una correlación positiva moderada entre el contenido de alcohol y la calidad del vino (aproximadamente 0.48). Esto sugiere que, en general, los vinos con un mayor contenido de alcohol tienden a recibir mejores calificaciones de calidad.

2. **Correlación entre Sulfatos y Calidad:**
   - Los sulfatos también muestran una correlación positiva moderada con la calidad del vino (aproximadamente 0.25). Esto indica que niveles más altos de sulfatos pueden estar asociados con una mejor calidad del vino.

3. **Correlación entre Acidez Volátil y Calidad:**
   - La acidez volátil presenta una correlación negativa moderada con la calidad del vino (aproximadamente -0.39). Esto sugiere que vinos con mayor acidez volátil tienden a tener calificaciones de calidad más bajas.

4. **Correlacion entre ácido cítrico y la acidez fija**

## **Importación de librerías y Carga del DATASET**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn import metrics

# Cargar el archivo CSV
url = 'https://raw.githubusercontent.com/tdiego74/Clase_Data_Science_I_PBLC/main/winequality-red.csv'
df = pd.read_csv(url)

# Traducción del nombre de las columnas
df = df.rename(columns={"fixed acidity": "Acidez Fija", "volatile acidity": "Acidez Volatil", "citric acid": "Ácido cítrico",
                        "residual sugar": "Azúcar residual", "chlorides": "Cloruros", "free sulfur dioxide": "Dióxido de azufre libre",
                        "total sulfur dioxide": "Dióxido de azufre total", "density": "Densidad", "sulphates": "Sulfatos",
                        "alcohol": "Alcohol", "quality": "Calidad"})
```

## **Separación de caracterísiticas y objetivo**

```python
# Separar las características (X) y el objetivo (y)
X = df.drop('Calidad', axis=1)
y = df['Calidad']
```

## **Dividir los datos en conjuntos de entrenamiento y prueba**

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## **Escalar los datos**

```python
# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## **Reducción de la dimensionalidad**

```python
# Aplicar PCA para reducir la dimensionalidad
pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

##**Entrenamiento del modelo REGRESIÓN LOGÍSTICA**

```python
# Entrenar el modelo de regresión logística
logistic_model = LogisticRegression(random_state=42, max_iter=10000)
logistic_model.fit(X_train_pca, y_train)
```

##**Predicciones sobre el modelo entrenado utilizando los conjuntos de prueba**

```python
# Hacer predicciones
y_pred_logistic = logistic_model.predict(X_test_pca)
```

##**Cálculo de métricas para evaluar el modelo**

```python
# Calcular métricas
accuracy_logistic = round(accuracy_score(y_test, y_pred_logistic), 2)
precision_logistic = round(precision_score(y_test, y_pred_logistic, average='weighted', zero_division=0), 2)
recall_logistic = round(recall_score(y_test, y_pred_logistic, average='weighted', zero_division=0), 2)
f1_logistic = round(f1_score(y_test, y_pred_logistic, average='weighted', zero_division=0), 2)

print(f"Regresión Logística - Accuracy: {accuracy_logistic}")
print(f"Regresión Logística - Precision: {precision_logistic}")
print(f"Regresión Logística - Recall: {recall_logistic}")
print(f"Regresión Logística - F1 Score: {f1_logistic}")
```
