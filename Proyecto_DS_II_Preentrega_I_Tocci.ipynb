#**Proyecto Calidad Vino Tinto**


## **Abstract**
El análisis de la calidad del vino tinto es esencial para entender cómo las propiedades fisicoquímicas del vino influyen en su percepción de calidad. El dataset "Wine Quality" contiene información sobre varios atributos químicos de muestras de vino tinto, así como una calificación de calidad asignada por expertos en una escala de 0 a 10. Este proyecto tiene como objetivo explorar las relaciones entre estas características y la calidad del vino, identificar patrones significativos y formular hipótesis que puedan ser útiles para productores y consumidores de vino.

## **Objetivo del Proyecto**
Este proyecto tiene como objetivo identificar los factores fisicoquímicos que influyen en la calidad del vino tinto, explorar las relaciones entre estas características y la calidad del vino, identificar patrones significativos y proporcionar recomendaciones prácticas que puedan ser útiles para productores y consumidores de vino con el fin de mejorar la calidad del producto.

Comprender los factores que afectan la calidad del vino puede ayudar a los productores a ajustar sus procesos de producción para satisfacer mejor las expectativas del mercado y aumentar sus ventas.

##**Descripción de la temática de los datos:**

El dataset proporcionado es sobre la calidad del vino tinto, que típicamente incluye variables relacionadas con sus características fisicoquímicas y una variable objetivo que indica la calidad del vino.

##**Descripción de las variables:**

**Acidez fija:** Componente principal de los ácidos en el vino que contribuye a su sabor y estabilidad.

**Acidez volátil:** Ácidos que contribuyen a un sabor avinagrado, influenciando negativamente la calidad si están en niveles altos.

**Ácido cítrico:** Contribuye a la frescura y a un sabor ácido en el vino.

**Azúcar residual:** Cantidad de azúcar que queda después de la fermentación, influenciando el dulzor del vino.

**Cloruros:** Contribuyen a la sensación de salinidad en el vino.

**Dióxido de azufre libre:** Previene el crecimiento microbiano y la oxidación.

**Dióxido de azufre total:** Suma del dióxido de azufre libre y el combinado, es crucial para la preservación del vino.

**Densidad:** Influencia el cuerpo del vino y está relacionado con el contenido de alcohol y azúcar.

**pH:** Medida de acidez/alcalinidad del vino, influye en la estabilidad y el sabor.

**Sulfatos:** Contribuyen a la conservación y ayudan a incrementar la intensidad del sabor.

**Alcohol:** Influencia el cuerpo, sabor y estabilidad del vino.

**Calidad:** Variable objetivo que mide la calidad del vino en una escala (generalmente de 0 a 10).

```python
#Categorizamos la Calidad en 3 niveles: Baja(entre 0 y 4), Media (entre 5 y 6) y Alta (entre 7 y 10)

bins = [0, 4, 6, 10]
labels = ['Baja', 'Media', 'Alta']
df['Rango de Calidad'] = pd.cut(df['Calidad'], bins=bins, labels=labels)
df.head()
```

####**Pregunta**: ¿Existe una correlación positiva entre el contenido de alcohol y la calidad del vino?

```python
# Scatterplot: Relación entre Alcohol y Calidad del Vino (Bivariado)
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(df['Alcohol'], df['Calidad'], alpha=0.5)
ax.set_title('Relación entre Alcohol y Calidad del Vino')
ax.set_xlabel('Alcohol')
ax.set_ylabel('Calidad')
plt.grid(False)
plt.show()
```

```python
# Scatterplot: Relación entre Alcohol y Calidad del Vino (Bivariado)
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(df['Alcohol'], df['Rango de Calidad'], alpha=0.5)
ax.set_title('Relación entre Alcohol y Calidad del Vino')
ax.set_xlabel('Alcohol')
ax.set_ylabel('Calidad')
plt.grid(False)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['Rango de Calidad'] = pd.Categorical(df['Rango de Calidad'], categories=['Alta', 'Media', 'Baja'], ordered=True)

# Invierto el código de categoría para que coincida con la posición visual deseada
df['Cod_Rango_Calidad'] = df['Rango de Calidad'].cat.codes
df['Cod_Rango_Calidad'] = df['Cod_Rango_Calidad'].replace({0: 2, 1: 1, 2: 0})

# Crear el scatter plot
fig, ax = plt.subplots(figsize=(8, 6))
scatter = ax.scatter(df['Alcohol'], df['Cod_Rango_Calidad'], alpha=0.5)
ax.set_title('Relación entre Alcohol y Calidad del Vino')
ax.set_xlabel('Alcohol')
ax.set_ylabel('Calidad')

# Ajustar las etiquetas del eje y para que coincidan con las categorías de calidad
ax.set_yticks([0, 1, 2])
ax.set_yticklabels(['Baja', 'Media', 'Alta'])

plt.grid(False)
plt.show()
```

```python
# Boxplot: Comparación de la Distribución del Alcohol entre Diferentes Calidades (Bivariado)
fig, ax = plt.subplots(figsize=(8, 6))
sns.boxplot(x='Rango de Calidad', y='Alcohol', data=df)
ax.set_title('Distribución del Alcohol por Calidad del Vino')
ax.set_xlabel('Calidad')
ax.set_ylabel('Alcohol')
plt.grid(False)
plt.show()
```


## **Interpretación:**
Los gráficos de dispersión y boxplot muestran que, en general, los vinos con mayor contenido de alcohol tienden a tener una mejor calidad.

Se observa una ligera tendencia positiva entre el contenido de alcohol y la calidad del vino.

## **Conclusión Hipotesis 1**
La hipótesis de que los vinos con mayor contenido de alcohol tienden a tener una mejor calificación de calidad es aceptada.

```python
# Boxplot: Comparación de la Distribución de la Acidez Volátil entre Diferentes Calidades (Bivariado)
fig, ax = plt.subplots(figsize=(8, 6))
sns.boxplot(x='Rango de Calidad', y='Acidez Volatil', data=df, ax=ax)
ax.set_title('Distribución de la Acidez Volátil por Calidad del Vino')
ax.set_xlabel('Calidad')
ax.set_ylabel('Acidez Volátil')
plt.grid(False)
plt.show()
```

## **Pregunta e Hipótesis 3**
####**Pregunta**: ¿Existe una correlación positiva entre los niveles de sulfatos y pH y la calidad del vino?
####**Hipótesis 3a**: La calidad del vino está positivamente correlacionada con niveles moderados de sulfatos.
####**Hipótesis 3b**: La calidad del vino está positivamente correlacionada con niveles moderados de pH.

```python
# Scatterplot: Relación entre Sulfatos y Calidad del Vino, con color representando pH (Multivariado)
fig, ax = plt.subplots(figsize=(8, 6))
scatter = ax.scatter(df['Sulfatos'], df['Cod_Rango_Calidad'], c=df['pH'], cmap='viridis', alpha=0.5)
ax.set_title('Relación entre Sulfatos y Calidad del Vino, con color representando pH')
ax.set_xlabel('Sulfatos')
ax.set_ylabel('Calidad')
plt.colorbar(scatter, label='pH')

# Ajustar las etiquetas del eje y para que coincidan con las categorías de calidad
ax.set_yticks([0, 1, 2])
ax.set_yticklabels(['Baja', 'Media', 'Alta'])

plt.grid(False)
plt.show()
```

```python
# Pairplot: Relación entre varias características fisicoquímicas y Calidad (Multivariado)
sns.pairplot(df, vars=['Alcohol', 'Acidez Volatil', 'Sulfatos', 'pH'], hue='Rango de Calidad', palette='viridis', diag_kind='kde')
plt.suptitle('Relación entre varias características fisicoquímicas y Calidad', y=1.02)
plt.show()
```

## **Interpretación:**
Los gráficos muestran que los niveles de sulfatos y pH están relacionados con la calidad del vino.

Los vinos con niveles moderados de sulfatos y pH tienden a tener una mejor calidad.

## **Conclusión Hipótesis 3**
La hipótesis de que la calidad del vino está positivamente correlacionada con niveles moderados de sulfatos y pH es aceptada.

### **Visualización de correlaciones**

```python
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linecolor='white', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.grid(False)
plt.show()
```

### **Observaciones Basadas en la Matriz de Correlación**


1. **Correlación entre Alcohol y Calidad:**
   - Existe una correlación positiva moderada entre el contenido de alcohol y la calidad del vino (aproximadamente 0.48). Esto sugiere que, en general, los vinos con un mayor contenido de alcohol tienden a recibir mejores calificaciones de calidad.

2. **Correlación entre Sulfatos y Calidad:**
   - Los sulfatos también muestran una correlación positiva moderada con la calidad del vino (aproximadamente 0.25). Esto indica que niveles más altos de sulfatos pueden estar asociados con una mejor calidad del vino.

3. **Correlación entre Acidez Volátil y Calidad:**
   - La acidez volátil presenta una correlación negativa moderada con la calidad del vino (aproximadamente -0.39). Esto sugiere que vinos con mayor acidez volátil tienden a tener calificaciones de calidad más bajas.

4. **Correlacion entre ácido cítrico y la acidez fija**
   - El ácido cítrico y la acidez fija muestran una correlación positiva moderada (aproximadamente 0.67), lo que indica que estos componentes a menudo varían juntos.

### **Interpretación**

La matriz de correlación confirma algunas de nuestras hipótesis iniciales y proporciona una base para el análisis posterior:

- La relación positiva entre el contenido de alcohol y la calidad del vino sugiere que los vinos con mayor contenido de alcohol tienden a ser mejor valorados.
- La relación negativa entre la acidez volátil y la calidad del vino refuerza la idea de que niveles más altos de acidez volátil pueden afectar negativamente la percepción de la calidad del vino.
- Las correlaciones entre las variables fisicoquímicas indican que existen patrones en cómo estos componentes están presentes en el vino, lo cual puede ser relevante para ajustar los procesos de producción y mejorar la calidad del producto.

Estas observaciones pueden ayudar a los productores de vino a centrarse en las características más influyentes y ajustar sus procesos de producción para mejorar la calidad del vino.

## **Recomendaciones Basadas en los Insights Observados**

1. **Optimización del Contenido de Alcohol**: Dado que los vinos con mayor contenido de alcohol tienden a tener mejor calidad, los productores de vino deberían considerar optimizar el contenido de alcohol para mejorar la calidad del vino.
2. **Control de la Acidez Volátil**: Los vinos con menor acidez volátil tienden a tener una mejor calidad. Se recomienda controlar y reducir los niveles de acidez volátil durante el proceso de producción.
3. **Balance de Sulfatos y pH**: Mantener niveles moderados de sulfatos y pH puede contribuir a una mejor calidad del vino. Los productores deberían monitorear y ajustar estos niveles para optimizar la calidad.
